## Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
movies = pd.read_csv("C:\\Users\\Asus\\Desktop\\ml-25m\\movies.csv")
ratings = pd.read_csv("C:\\Users\\Asus\\Desktop\\ml-25m\\ratings.csv")
joined = ratings.merge(movies, on='movieId', how='left')
joined
```

```python
five = joined[joined.rating == 5].groupby(["movieId","title"]).aggregate({"rating":"count"})[0:1]
five
```

## Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv("power.csv")
power 
```

```python
countries = ["Latvia", "Lithuania", "Estonia"]
categories = ["4", "12", "21"]

baltic = power.loc[(power["country"].isin(countries)) & (power["category"].isin(categories)) & ((power["year"] < 2010) & (power["year"] > 2005)) & (power["quantity"] > 0)] 
baltic
```
